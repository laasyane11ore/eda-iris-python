```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset from Seaborn
iris = sns.load_dataset("iris")

# Display basic information about the dataset
print(iris.info())

# Display the first few rows of the dataset
print(iris.head())

# Summary statistics
print(iris.describe())

# Pairplot to visualize relationships between features
sns.pairplot(iris, hue="species")
plt.show()

# Box plot to visualize distribution of features by species
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length by Species")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Univariate Analysis: Histograms for each feature
plt.figure(figsize=(12, 8))

# Iterate through each feature and create a separate subplot for each
for i, column in enumerate(iris.columns[:-1], start=1):  # Exclude the 'species' column
    plt.subplot(2, 2, i)  # Create a 2x2 grid of subplots
    sns.histplot(iris[column], kde=True)
    plt.title(f"Univariate Analysis: {column.capitalize()} Histogram")

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis
# Pairplot colored by species
plt.figure(figsize=(10, 6))
sns.pairplot(iris, hue="species")
plt.title("Bivariate Analysis: Pairplot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
pairplot = sns.pairplot(iris, diag_kind="kde")
pairplot.fig.suptitle("Multivariate Analysis: Scatterplot Matrix", y=1.02)
plt.show()
```
